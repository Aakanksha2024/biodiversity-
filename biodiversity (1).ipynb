```python
# Load species_info.csv from a specific directory
species_info = pd.read_csv('C:/Users/aakan/OneDrive/Desktop/species_info.csv')

# Load observations.csv from a specific directory
observations = pd.read_csv('C:/Users/aakan/OneDrive/Desktop/observations.csv')
```

```python
# Print the column names of species_info
print("Column names of species_info:")
print(species_info.columns)

# Print the column names of observations
print("\nColumn names of observations:")
print(observations.columns)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# plotting the distribution of conservation_status for animals 
plt.figure(figsize=(10, 6))
sns.countplot(data=species_info, x='conservation_status')
plt.title('Distribution of Conservation Status for Animals')
plt.xlabel('Conservation Status')
plt.ylabel('Count')
plt.show()
```

```python
#plot the distribution of species types by conservation status
plt.figure(figsize=(10, 6))
sns.countplot(data=species_info, x='conservation_status', hue='category')
plt.title('Conservation Status by Species Type')
plt.xlabel('Conservation Status')
plt.ylabel('Count')
plt.show()
```

```python
# Create a stacked bar plot to show the relationship between species types and conservation status
ct = pd.crosstab(species_info.category, species_info.conservation_status)
ct.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Conservation Status by Species Type')
plt.xlabel('Species Type')
plt.ylabel('Count')
plt.show()
```

```python
# Identify the most spotted species at each park
most_spotted_species = observations.groupby('park_name')['scientific_name'].agg(lambda x:x.value_counts().index[0])
print("Most spotted species at each park:")
print(most_spotted_species)
```
